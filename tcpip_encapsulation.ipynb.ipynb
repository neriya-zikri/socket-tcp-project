```python
# TCP/IP Encapsulation Project

This notebook simulates TCP/IP encapsulation using application-layer messages from a CSV file.
```

```python
import pandas as pd

csv_path = "./data/group05_http_input.csv"
messages_df = pd.read_csv(csv_path)
messages_df.head()
```

```python
## TCP/IP Encapsulation Simulation

Each application message from the CSV file represents Application Layer data.
This data is encapsulated inside a TCP segment (Transport Layer),
which is then encapsulated inside an IP packet (Internet Layer).
```

```python
for index, row in messages_df.iterrows():
    print(f"Message {row['msg_id']}:")
    print("Application Layer:", row['message'])
    print("Transport Layer: TCP Segment")
    print("Internet Layer: IP Packet")
    print("-" * 40)
```
